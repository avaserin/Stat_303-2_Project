### Data cleaning
*By Emily Zhang*

From the data quality check we realized that:

1. We renamed `model year` and `car name` because these columns have spaces in them. We renamed them `model_year` and `car_name` so we can better call on these columns. 

2. We binned the `cylinder` column with specific bins that we see from the distribution above.

3. There were some values in the `horsepower` column that were '?'. We replaced these with NaN to avoid errors in the code. 

4. We changed the `horespower` column to be type float.

5. we mapped the `origin` column to country string names from key (found in Kaggle).

The code below implements the above cleaning.

```python
data.head()
```

```python
# Renaming model year and car name column
data = data.rename(columns={'model year': 'model_year'})
data = data.rename(columns={'car name': 'car_name'})
```

```python
# Binning cylinder column
bins = [3,4,5,6,8]
data['cyl_binned'] = pd.cut(data['cylinders'],bins = bins)
```

```python
# Replace '?' with 'na'
data = data.replace('?', np.nan)
```

```python
# Convert columns from object to float datatype
data['horsepower'] = data['horsepower'].astype(float)
```

```python
# Mapping origin column to countries from key
data['origin'] = data['origin'].map({1: 'US', 2: 'Europe', 3: 'Asia'})
```

```python
# Split the data into training and test
np.random.seed(2)
train = data.sample(round(data.shape[0]*0.8)) # 80% of data in training set
test = data.drop(train.index)
```

```python
# VIF to detect multicollinearity
X = train[['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model_year']]

from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

# Check for missing values and convert data to numeric type
X = X.apply(pd.to_numeric, errors='coerce')

# Drop rows with missing values
X = X.dropna()

X = add_constant(X)
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns

for i in range(len(X.columns)):
    vif_data.loc[i,'VIF'] = variance_inflation_factor(X.values, i)

print(vif_data)
```

#### Model with all first-order predictors

```python
model1 = smf.ols(formula = 'mpg ~ cyl_binned + horsepower + weight + acceleration + model_year', data = train).fit()
model1.summary()
```

```python
pred_mpg = model1.predict(test)
print("RMSE:", np.sqrt(((test.mpg - pred_mpg)**2).mean()))
```

#### Model with quadratic term for weight

```python
model2 = smf.ols(formula = 'mpg ~ cyl_binned + horsepower + weight + I(weight**2) + acceleration + model_year', data = train).fit()
model2.summary()
```

```python
pred_mpg = model2.predict(test)
print("RMSE:", np.sqrt(((test.mpg - pred_mpg)**2).mean()))
```

#### Model with quadratic terms for horsepower and weight

```python
model3 = smf.ols(formula = 'mpg ~ horsepower + I(horsepower**2) + weight + I(weight**2) + acceleration + model_year', data = train).fit()
model3.summary()
```

```python
pred_mpg = model3.predict(test)
print("RMSE:", np.sqrt(((test.mpg - pred_mpg)**2).mean()))
```

#### Model with mpg logged and above quadratic transformations

```python
model4 = smf.ols(formula = 'np.log(mpg) ~ horsepower + I(horsepower**2) + weight + I(weight**2) + acceleration + model_year', data = train).fit()
model4.summary()
```

```python
pred_mpg = np.exp(model4.predict(test))
np.sqrt(((test.mpg - pred_mpg)**2).mean())
```

```python
#Plotting residuals vs fitted values
sns.set(rc={'figure.figsize':(10,6)})
sns.scatterplot(x = (model4.fittedvalues), y=(model4.resid),color = 'orange')
sns.lineplot(x = [model4.fittedvalues.min(),model4.fittedvalues.max()],y = [0,0],color = 'blue')
plt.xlabel('Fitted values')
plt.ylabel('Residuals')
```

```python
#Studentized residuals
out = model4.outlier_test()

#Plotting studentized residuals vs fitted values
sns.scatterplot(x = (model4.fittedvalues), y=(out.student_resid),color = 'orange')
sns.lineplot(x = [model4.fittedvalues.min(),model4.fittedvalues.max()],y = [0,0],color = 'blue')
plt.xlabel('Fitted values')
plt.ylabel('Studentized Residuals')
```

```python
# Number of points with absolute studentized residuals greater than 3
np.sum((np.abs(out.student_resid)>3))
```

```python
# Computing the leverage statistic for each observation
influence = model4.get_influence()
leverage = influence.hat_matrix_diag
```

```python
# Visualizng leverage against studentized residuals
sns.set(rc={'figure.figsize':(15,8)})
sm.graphics.influence_plot(model4);
```

```python
out = model4.outlier_test()
average_leverage = (model4.df_model+1)/model4.nobs
high_leverage_threshold = 4*average_leverage
print("Number of high leverage points:", np.sum(leverage>high_leverage_threshold))
```
